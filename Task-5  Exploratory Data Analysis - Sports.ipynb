#### and now we will find the factors contributing win or loss of a team. 

```python
matche_cleaned.columns
```

```python
import statsmodels.api as sm

from sklearn.preprocessing import LabelEncoder
l= LabelEncoder()

team1=l.fit_transform(matche_cleaned["team1"])
team2=l.fit_transform(matche_cleaned["team2"])
toss_winner=l.fit_transform(matche_cleaned["toss_winner"])
toss_decision=l.fit_transform(matche_cleaned["toss_decision"])
winner=l.fit_transform(matche_cleaned["winner"])
player_of_match=l.fit_transform(matche_cleaned["player_of_match"])
venue=l.fit_transform(matche_cleaned["venue"])
result=l.fit_transform(matche_cleaned["result"])

matche_cleaned_new = matche_cleaned.drop(['team1','team2','toss_winner','toss_decision','winner',
                             'player_of_match','player_of_match','venue','result'],axis=1)

matche_cleaned_new['team1']=team1
matche_cleaned_new['team2']=team2
matche_cleaned_new['toss_winner']=toss_winner
matche_cleaned_new['toss_decision']=toss_decision
matche_cleaned_new['winner']=winner
matche_cleaned_new['player_of_match']=player_of_match
matche_cleaned_new['venue']=venue
matche_cleaned_new['result']=result

matche_cleaned_new.head(3)

x= matche_cleaned_new.drop(['result'],axis=1)
y=matche_cleaned_new["result"]
X_1 = sm.add_constant(x)
#Fitting sm.OLS model
model = sm.OLS(y,X_1).fit()
model.pvalues.round(2)
```

##### win_by_runs and win_by_wickets are  the factors contributing win or loss of a team.
